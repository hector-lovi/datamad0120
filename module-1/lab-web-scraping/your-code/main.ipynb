####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url7 = 'https://www.emsc-csem.org/Earthquake/'

res7 = requests.get(url7)
soup7 = BeautifulSoup(res7.text, 'html.parser')
# print(soup7.prettify)
````

````python
def procesaFila(tag):
    m = tag.find_all("td")
    #print(m[1])
    return {
        "Date":re.findall(r'(\d\d\d\d-\d\d-\d\d)', (m[3].text)),
        "Hour":re.findall(r'(\d\d:\d\d:\d\d)', (m[3].text)),
        "Latitude":m[4].text + m[5].text,
        "Longitude":m[6].text + m[7].text,
        "Depth":m[8].text,
        "Magnitude":m[10].text,
        "Region":m[11].text
    }
    
earthquakes = soup7.find_all('tbody')[0]
earthquakes_dict = [procesaFila(fila) for fila in earthquakes.find_all("tr")[1:21]]
````

````python
df = pd.DataFrame(earthquakes_dict)
df
````
